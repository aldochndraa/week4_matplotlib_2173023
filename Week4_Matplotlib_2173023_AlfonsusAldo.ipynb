```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
data = [25, 45, 55, 125, 225]
kategori = ['A','B','C','D','E']
```

```python
plt.bar(kategori,data)
plt.grid()

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

```python
plt.bar(kategori,
        data,
        color='red',
        alpha=0.25)

plt.grid(linestyle='--',
         linewidth=2,
         axis='y',
         alpha=0.50)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

```python
plt.barh(kategori,data)

plt.grid(linestyle='--',
         linewidth=1,
         axis='x',
         alpha=0.75)

plt.xlabel('Jumlah')
plt.ylabel('Kategori')
plt.title('Contoh Horizontal Bar Plot')

plt.show()
```

```python
data1 = [25, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ['A','B','C','D','E']
```

```python
plt.bar(kategori, data1,label='Data 1')
plt.bar(kategori, data2,label='Data 2', bottom=data1)

plt.legend()

plt.grid(linestyle='--',
         linewidth=1,
         axis='y',
         alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Stacked Bar Plot')

plt.show()
```

```python
data1 = [25, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ['A','B','C','D','E']

x=np.arange(len(kategori))
x
```

```python
width=0.35

plt.bar(x-width/2, data1, width,label='Data 1')
plt.bar(x+width/2, data2, width,label='Data 2')

plt.xticks(x, kategori)

plt.grid(linestyle='--',
         linewidth=1,
         axis='y',
         alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Grouped Bar Plot')

plt.legend()
plt.show()
```

```python
kategori = ['A','B','C','D','E']
data = [215,130,245,210,225]
```

```python
plt.pie(data,
        labels=kategori,
        autopct='%1.1f%%',
        startangle=90)

plt.title('Contoh Pie Plot')

plt.show()
```

```python
kategori = ['A','B','C','D','E']
data = [215,130,245,210,225]
warna = ['pink','cyan','#e03364','yellowgreen','skyblue']
```

```python
plt.pie(data,
        labels=kategori,
        colors=warna,
        autopct='%1.1f%%',
        startangle=90)

plt.legend()
plt.title('Contoh Pie Plot')

plt.show()
```

```python
kategori = ['A','B','C','D','E']
data = [215,130,245,210,225]
warna = ['pink','cyan','#e03364','yellowgreen','skyblue']
explode_var=[0.,0.,0.2,0.,0.08]
```

```python
plt.pie(data,
        labels=kategori,
        colors=warna,
        explode=explode_var,
        autopct='%1.1f%%',
        startangle=90)

plt.title('Contoh Pie Plot')

plt.show()
```
